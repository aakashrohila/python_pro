```python
class Node:

  def __init__(self,data):
    self.data = data
    self.next = None

class Linked_list:

  def __init__(self):
    self.start = None
#1
  def insert_First(self,data):
    node = Node(data)

    if self.start == None:
      self.start = node
      node.data = data
      node.next = None

    else:
      node.next = self.start
      self.start = node
#2
  def insert_Last(self,data):

    node = Node(data)

    if self.start == None:
      self.start = node
      node.data = data
      node.next = None

    else:
      trav = self.start
      
      while trav.next != None :
        trav = trav.next

      trav.next = node
#3
  def Delete_right(self):

    trav = self.start

    if trav == None:
      pass
    
    elif trav.next == None:
      print("Deleting Data",trav.data)
      del trav
      trav = None
      self.start = None
 
    else:
      while trav.next.next != None:
        trav = trav.next

      print("Deleting Data",trav.next.data)
      del trav.next
      trav.next = None
#4
  def Delete_left(self):

    trav = self.start

    if trav == None:
      pass

    else:
      print("Deleting Data",trav.data)
      self.start = trav.next
      del trav

  def Print_nodes(self):

    if self.start == None:
      print("Empty")
    
    else:
      trav = self.start

      while trav != None:
        print(trav.data,"->",end='')
        trav = trav.next
      print()
```

```python
llist = Linked_list()
while True:
  print("Enter your choice")
  print("1.Insert Left")
  print("2.Insert Right")
  print("3.Delete Left")
  print("4.Delete Right")
  print("5.Print Linked List")
  print("0. Stop ")

  
  choice = int(input("Choice: "))

  if choice == 1:
    data = input("Enter your data: ")
    llist.insert_First(data)

  elif choice==2:
    data = input("Enter your data: ")
    llist.insert_Last(data)

  elif choice==3:
    llist.Delete_left()

  elif choice==4:
    llist.Delete_right()

  elif choice==5:
    llist.Print_nodes()

  else:
    break

```
